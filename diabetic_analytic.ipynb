```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Read the data present in the TSV File
df = pd.read_csv("diabetese.tsv", sep="\t")#sep=seperator, \t means files are seperated by tab character
df.head()
```

```python
df.info()
```

```python
#Avg age of diabetic patients
mean_age = df['AGE'].mean()
print(f"Mean Age is : {mean_age}")
```

```python
#Variance
var_age = df['AGE'].var()
print(f"Variance in age : {var_age}")
```

```python
#Standard Deviation
sd_age = df['AGE'].std()
print(f"Standard Deviation in age : {sd_age}")
```

```python
#Boxplot for age field
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'], showmeans=True)
plt.grid(color='gray', linestyle = 'dotted')
plt.xlabel("Age")
plt.ylabel("Years")
plt.show()
```

```python
#Boxplotting for BMI
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'], vert=False, showmeans=True)
plt.grid(color='gray', linestyle='dotted')
plt.xlabel('Value')
plt.ylabel('BMI')
plt.show()
```

```python
#Boxplotting for BP
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'], vert=False, showmeans=True)
plt.grid(color='gray', linestyle='dotted')
plt.xlabel('Value')
plt.ylabel('BP')
plt.show()
```

```python
#Creating histograms
df['AGE'].hist(bins=12)
plt.title('Age distribution of diabetic patients')
plt.xlabel('Age Range')
plt.ylabel('Count')
```

```python
df['BMI'].hist(bins=12)
plt.title('BMI distribution of diabetic patients')
plt.xlabel('BMI Range')
plt.ylabel('Count')
```

```python
hours_studies = [2,3,1,4,2,5,4,6,3,5]
exam_scores = [70,80,60,85,75,90,88,92,80,95]
plt.scatter(hours_studies, exam_scores, color='blue', marker='+') #marker default = 'o'
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Scatter plot: Hours Studied vs Exam Score')
plt.show()
```

```python
plt.scatter(df['BMI'], df['Y'], color='blue')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('BMI vs Disease Progression')
plt.show()
```

```python
plt.scatter(df['S5'], df['Y'], color='blue')
plt.xlabel('S5')
plt.ylabel('Disease Progression')
plt.title('S5 vs Disease Progression')
plt.show()
```
